```python
ROOT = "/u/amo-d0/ugrad/connor/scratch/data/pascal_voc/VOC2012/"
```

```python
CSV_ROOT = ROOT+"ImageSets/Segmentation/"

train_path = CSV_ROOT+"train.txt"
val_path = CSV_ROOT+"val.txt"
```

```python
train_ids = open(train_path).readlines()
train_ids = list(map(str.strip, train_ids))

val_ids = open(val_path).readlines()
val_ids = list(map(str.strip, val_ids))

train_ids[:4], val_ids[:4]
```

```python
len(train_ids), len(val_ids)
```

```python
import skimage.io as imgio
import matplotlib.pyplot as plt
%matplotlib inline
plt.matplotlib.rcParams['figure.figsize'] = [14, 8]
```

```python
JPEG_PATTERN = ROOT+"JPEGImages/{}.jpg"
ANNO_PATTERN = ROOT+"SegmentationObject/{}.png"
```

```python
_, axs = plt.subplots(2, 4)

for idx, ax in zip(train_ids, axs.T):
    ax[0].imshow(imgio.imread(JPEG_PATTERN.format(idx)))
    ax[1].imshow(imgio.imread(ANNO_PATTERN.format(idx)))
```

```python
_, axs = plt.subplots(2, 4)

for idx, ax in zip(val_ids, axs.T):
    ax[0].imshow(imgio.imread(JPEG_PATTERN.format(idx)))
    ax[1].imshow(imgio.imread(ANNO_PATTERN.format(idx)))
```
